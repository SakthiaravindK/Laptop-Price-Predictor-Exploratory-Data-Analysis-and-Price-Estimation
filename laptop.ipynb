```python
#Inferential Statistics
#Independent Two-Sample t-Test
from scipy.stats import ttest_ind

# Filter data for 8GB and 16GB RAM
price_8gb = laptop_df[laptop_df['Ram'] == 8]['Price']
price_16gb = laptop_df[laptop_df['Ram'] == 16]['Price']

# Perform t-test
t_stat, p_value = ttest_ind(price_8gb, price_16gb, equal_var=False)

print(f"T-statistic: {t_stat}, P-value: {p_value}")
#decision:
alpha=0.05
if p_value<alpha :
    print("reject the null hupothesis")
else:
    print("fails to reject the null hypothesis")
```

```python
# Chi-Square Test of Independence
from scipy.stats import chi2_contingency

# Create a contingency table
contingency_table = pd.crosstab(laptop_df['Company'], laptop_df['OpSys'])

# Perform Chi-Square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

print(f"Chi2 Statistic: {chi2}, P-value: {p}")
#decision:
alpha=0.05
if p_value<alpha :
    print("reject the null hupothesis")
else:
    print("fails to reject the null hypothesis")
```

```python
import pandas as pd

# Load your dataset
laptop_df = pd.read_csv('laptopData.csv')  # Replace with your actual file path

# Calculate Q1 (25th percentile) and Q3 (75th percentile) for the 'Price' column
Q1 = laptop_df['Price'].quantile(0.25)
Q3 = laptop_df['Price'].quantile(0.75)
IQR = Q3 - Q1  # Interquartile Range

# Define the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter out the outliers
laptop_df_cleaned = laptop_df[(laptop_df['Price'] >= lower_bound) & (laptop_df['Price'] <= upper_bound)]

# Display the number of records before and after outlier removal
print(f"Original dataset size: {laptop_df.shape[0]}")
print(f"Cleaned dataset size: {laptop_df_cleaned.shape[0]}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Boxplot before handling outliers
plt.figure(figsize=(10, 5))
sns.boxplot(x=laptop_df['Price'])
plt.title('Boxplot of Laptop Prices (Before Handling Outliers)')
plt.show()

# Boxplot after removing outliers
plt.figure(figsize=(10, 5))
sns.boxplot(x=laptop_df_cleaned['Price'])
plt.title('Boxplot of Laptop Prices (After Removing Outliers)')
plt.show()
```
